```python
# intalize an embedding layer
vocab_size = 1000
embedding_data = 100
embedding = nn.Embedding(vocab_size, embedding_data)

# create some input indices
input_indices = torch.LongTensor([1,5,3,2])

# apply embedding layer
embedded_output = embedding(input_indices)

# the output will be a tensor of shape (4,100) where 4 is # inputs and 100 is the 
# dimensionality of the embdedding vector
print(embedded_output.shape)
```
